#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.

```python
temperatura_f = float(input("Digite a temperatura em Fahrenheit: "))
temperatura_c = (temperatura_f - 32) * 5 / 9

print(f"A temperatura em Celsius é: {temperatura_c:.2f}")
```

#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
ganho_hora = float(input("Qual o valor da sua hora de trabalho? "))
horas_trabalhadas = int(input("Quantas horas você ytrabalha por mês? "))

print(f'Você cobra {ganho_hora} por hora de trabalho e trabalha {horas_trabalhadas} horas por mês!')

print(ganho_hora * horas_trabalhadas)

```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salario_bruto = ganho_hora * horas_trabalhadas
print(f'O seu salário bruto é de R${salario_bruto:.2f}')
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
imposto_renda = salario_bruto * 0.11
print(f'O seu imposto de renda é de: R${imposto_renda:.2f}')
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
inss = salario_bruto * 0.08
print(f'O seu INSS é de: R${inss:.2f}')
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
sindicato = salario_bruto * 0.05
print(f'O seu sindicato é de: R${sindicato:.2f}')
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
salario_liquido = salario_bruto - imposto_renda - inss - sindicato
print(salario_bruto)
print(imposto_renda)
print(inss)
print(sindicato)

print(f'O seu salário líquido foi de: R${salario_bruto - imposto_renda - inss - sindicato:.2f}')
```

#### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
import math


lata_tinta_v = 80.00
lata_tinta_l = 18
cobertura_por_litro = 3
area = float(input("Digite a área a ser pintada em metros quadrados: "))

tinta_necessaria = area / cobertura_por_litro

latas_qtd = math.ceil(tinta_necessaria / lata_tinta_l)

print(f'A área total é de {area} metros quadrados. Você irá precisar de {latas_qtd} latas de tinta de 18L e custará R${lata_tinta_v * latas_qtd}!')
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tamanho = float(input("Digite o tamanho do arquivo em MB: "))
velocidade = float(input("Digite a velocidade da internet em Mbps: "))

tamanho_em_megabits = tamanho * 8

tempo = tamanho_em_megabits / velocidade

print("O tempo estimado de download é de: ", tempo, "segundos")
```
